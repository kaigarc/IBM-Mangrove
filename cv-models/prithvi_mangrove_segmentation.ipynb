```python
import os
import torch
import matplotlib.pyplot as plt
import numpy as np
import rasterio
import yaml
from prithvi.Prithvi import MaskedAutoencoderViT

NO_DATA = -9999
NO_DATA_FLOAT = 0.0001
PERCENTILES = (0.1, 99.9)
```

```python
def load_raster(path, crop=None):
    with rasterio.open(path) as src:
        img = src.read()

        # load first 6 bands
        img = img[:6]

        img = np.where(img == NO_DATA, NO_DATA_FLOAT, img)
        if crop:
            img = img[:, -crop[0]:, -crop[1]:]
    return img

def enhance_raster_for_visualization(raster, ref_img=None):
    if ref_img is None:
        ref_img = raster
    channels = []
    for channel in range(raster.shape[0]):
        valid_mask = np.ones_like(ref_img[channel], dtype=bool)
        valid_mask[ref_img[channel] == NO_DATA_FLOAT] = False
        mins, maxs = np.percentile(ref_img[channel][valid_mask], PERCENTILES)
        normalized_raster = (raster[channel] - mins) / (maxs - mins)
        normalized_raster[~valid_mask] = 0
        clipped = np.clip(normalized_raster, 0, 1)
        channels.append(clipped)
    clipped = np.stack(channels)
    channels_last = np.moveaxis(clipped, 0, -1)[..., :3]
    rgb = channels_last[..., ::-1]
    return rgb
```

```python
import rasterio
from rasterio.transform import from_origin
import tifffile

def save_image_as_tif(image_array, output_path):
    """
    Save an image array to a TIFF file.

    Parameters:
    - image_array: NumPy array of the image, expected to be in channel-first format (channels, height, width).
    - output_path: String specifying the path where the TIFF file will be saved.
    """
    # Define the transform; adjust parameters as needed for your data
    transform = from_origin(0, 0, 1, 1)  # Example: top-left corner at (0, 0), pixel size (1, 1)

    with rasterio.open(
        output_path, 'w',
        driver='GTiff',
        height=image_array.shape[1],
        width=image_array.shape[2],
        count=image_array.shape[0],
        dtype=image_array.dtype,
        transform=transform
    ) as dst:
        for i in range(image_array.shape[0]):
            # Rasterio expects 1-based indexing for bands
            dst.write(image_array[i], i + 1)
```

```python
output_dir_images = '/content/drive/MyDrive/IBM_Spring24/mangrove_dataset/jamaica_tif/images/'
output_dir_labels = '/content/drive/MyDrive/IBM_Spring24/mangrove_dataset/jamaica_tif/labels/'

def produce_tif(img_array, label_array):
    for i in range(len(img_array)):
        # Extract the first 6 channels and convert to channel-first format correctly
        rgb_image = img_array[i][:, :, 0:6]
        rgb_image_cf = np.moveaxis(rgb_image, -1, 0)  # cf stands for channel-first

        # Handle binary mask and create an RGB label without moving axis
        binary_mask = label_array[i]
        rgb_label = np.zeros((binary_mask.shape[0], binary_mask.shape[1], 3), dtype=np.uint8)
        rgb_label[binary_mask[:, :, 0] == 1] = [255, 255, 255]

        # Use randomization to keep/drop if no mangrove
        if (np.sum(binary_mask) == 0.0):
            number = random.random()
            if number <= 0.2:
                save_image_as_tif(rgb_image_cf, f"{output_dir_images}image_{i}.tif")
                save_image_as_tif(rgb_label, f"{output_dir_labels}label_{i}.tif")
        else:
            save_image_as_tif(rgb_image_cf, f"{output_dir_images}image_{i}.tif")
            save_image_as_tif(rgb_label, f"{output_dir_labels}label_{i}.tif")

# Example usage
produce_tif(images, labels)
```

```python
#generating train/test/val split (0.8, 0.1, 0.1)
import os

def list_tif_filenames(directory):
    """
    Scans through the specified directory and appends the name of each '.tif' file
    without the extension to a list.

    Parameters:
    - directory (str): Path to the directory to scan.

    Returns:
    - List[str]: A list of filenames without the '.tif' extension.
    """
    filenames = []
    # Loop through each file in the specified directory
    for file in os.listdir(directory):
        # Check if the file is a '.tif' file
        if file.endswith('.tif'):
            # Remove the '.tif' extension and add to the list
            filenames.append(os.path.splitext(file)[0])

    return filenames

# Usage example
directory_path1 = '/content/drive/MyDrive/IBM_Spring24/mangrove_dataset/jamaica_tif/images'
#directory_path2 = '/content/drive/MyDrive/IBM_Spring24/mangrove_dataset/jamaica_tif/labels'
image_filenames = list_tif_filenames(directory_path1)
#label_filenames = list_tif_filenames(directory_path2)
print(len(image_filenames))
#print(len(label_filenames))
```

```python
import random

def distribute_filenames(filenames):
    """
    Distributes filenames into train, test, and validation sets and writes them to respective files.

    Parameters:
    - filenames (list): List of filenames to distribute.
    """
    # Shuffle the list to ensure random distribution
    random.shuffle(filenames)

    # Calculate the indices for slicing the list
    total_count = len(filenames)
    train_count = int(0.8 * total_count)
    test_count = int(0.1 * total_count)
    val_count = total_count - (train_count + test_count)  # Ensures that all files are included

    # Slice the list to create train, test, and validation subsets
    train_filenames = filenames[:train_count]
    test_filenames = filenames[train_count:train_count + test_count]
    val_filenames = filenames[train_count + test_count:]

    # Write each subset to the corresponding file
    with open('train.txt', 'w') as f:
        for filename in train_filenames:
            filename = filename[0:-6]
            f.write(filename + '\n')

    with open('test.txt', 'w') as f:
        for filename in test_filenames:
            filename = filename[0:-6]
            f.write(filename + '\n')

    with open('val.txt', 'w') as f:
        for filename in val_filenames:
            filename = filename[0:-6]
            f.write(filename + '\n')

distribute_filenames(image_filenames)
```

```python
import os
import re

def rename_files_numeric(directory):
    """
    Renames each file in the specified directory by extracting the numeric part
    of each filename and formatting the new name as image_{number}_image.tif.

    Parameters:
    - directory (str): Path to the directory containing the files.
    """
    # Regular expression to extract numbers
    pattern = re.compile(r'\d+')

    # Iterate over all files in the directory
    for filename in os.listdir(directory):
        # Construct the full file path
        full_path = os.path.join(directory, filename)

        # Ensure it's a file and not a directory
        if os.path.isfile(full_path):
            # Find all numbers in the filename
            numbers = pattern.findall(filename)

            if numbers:
                # Assume the first number is what we want
                number = numbers[0]
                # Create the new filename
                new_filename = f"image_{number}_label.tif"
                new_file_path = os.path.join(directory, new_filename)

                # Rename the file
                os.rename(full_path, new_file_path)
                print(f"Renamed {full_path} to {new_file_path}")

# Example usage
directory_path = '/content/drive/MyDrive/IBM_Spring24/mangrove_dataset/jamaica_tif/labels'
rename_files_numeric(directory_path)
```
